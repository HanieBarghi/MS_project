```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.patches import Rectangle
```

```python
# False Positive throry
# generating data
p = 5000
th_max = 0.2
th_min = 0.004
n = 20
sigma = 0.007
beta = 0.05
s_0 = 20

th = th_min
fp = []
tp = []
xx = []
R_max = sigma * math.sqrt(s_0) * math.log(p) / n
sig_max = math.sqrt(math.pow(sigma, 2) * (1 + math.sqrt(math.log(p / n))))

while th <= th_max:
  if th <= R_max:
    temp1 = 1
  else:
    temp1 = 0
  temp2 = 1 - 0.5 * (math.erf((th + R_max)/(math.sqrt(2) * sig_max)) + math.erf((th - R_max)/(math.sqrt(2) * sig_max)))
  if temp1 >= temp2:
    fp.append(temp1)
  else:
    fp.append(temp2)
  
  if th <= (R_max + beta):
    temp1 = 1
  else:
    temp1 = 0
  temp2 = 1 - 0.5 * (math.erf((th + R_max + beta)/(math.sqrt(2) * sig_max)) + math.erf((th - R_max - beta)/(math.sqrt(2) * sig_max)))
  if temp1 <= temp2:
    tp.append(temp1)
  else:
    tp.append(temp2)
  xx.append(th)
  th += 0.000001
# ---------------------------------------------------------------------------------------------------------------------------------------  

def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(7.5, 4.5))    
    
    someX, someY = 0.02, 0.0
    currentAxis = plt.gca()
    currentAxis.add_patch(Rectangle((someX, someY), 0.04, 0.03, facecolor = "#0B9F2A"))
    ax.plot(xx, tp, '-', label = 'non-Zero Dimensions', color = '#0E6FEC')
    ax.plot(xx, fp, '-', label = 'Zero Dimensions', color = '#C23005')
    
    ax.grid(linestyle='--')
    
    plt.ylabel('Probability of Extracting an Arbitrary Feature')
    plt.xlabel('Threshold')

    plt.xscale('log')
    ax.legend()

demo('default')
```
